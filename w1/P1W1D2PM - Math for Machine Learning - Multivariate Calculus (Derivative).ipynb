## B. Derivative

Let's find the derivative from a function $y = f(x) = 2x^2 + 4x -1$

Before declaring the function $f(x)$, you need to first declare the variable $x$ as a mathematical variable, not a variable in Python. If the variable $x$ is not declared as a mathematical variable beforehand, an error like this will appear.

### B.1 - Define the Symbols and the Function

```python
# Define the mathematical symbols

x = sy.Symbol('x')
x
```

```python
# Define the function

f = 2*x**2 + 4*x - 1
f
```

### B.2 - First Derivative

```python
# First derivative - syntax 1

first_df_syntax_1 = f.diff()
first_df_syntax_1
```

```python
# First derivative - syntax 2

first_df_syntax_2 = sy.diff(f)
first_df_syntax_2
```

### B.3 - Second Derivative and Third Derivative

```python
# Second derivative - syntax 1

second_df_syntax_1 = f.diff().diff()
second_df_syntax_1
```

```python
# Second derivative - syntax 2

second_df_syntax_2 = sy.diff(sy.diff(f))
second_df_syntax_2
```

```python
# Third derivative - syntax 1

third_df_syntax_1 = f.diff().diff().diff()
third_df_syntax_1
```

```python
# Third derivative - syntax 2

third_df_syntax_2 = sy.diff(sy.diff(sy.diff(f)))
third_df_syntax_2
```

### B.4 - Slope between Points

First, let's create an array of $x$ and their result from a function $y = f(x) = 2x^2 + 4x - 1$

```python
# Create an array of x

X = np.linspace(-10,10, 100)

print('Shape : ', X.shape)
print('Value of X : \n', X)
```

```python
# Create the result of a function from an array of x

Y = 2*X**2+4*X-1

print('Shape : ', Y.shape)
print('Value of Y : \n', Y)
```

```python
# Plot the function

plt.plot(X, Y)
```

To find the slope between two points, we can use the formula as below :

$m = \frac {Δy} {Δx} = \frac {y2 - y1} {x2 - x1)}$


```python
# Find the slope between two points

slope = np.gradient(Y, X)

print('Shape of slope : ', slope.shape)
print('Slope : \n', slope)
```

```python
# Plot of the slope

plt.plot(X, slope)
```

## C. Critical Points

Let's find the critical points (global maximum and global minimum) from a function $y = f(x) = x^3 - 3x + 1$

### C.1 - Find Local/Global Minimum

```python
# Find the local/global minimum points

## Import library
from scipy.optimize import minimize_scalar

## Define the function
def func(x):
  return x**3 - 3*x + 1

## Display the result
result = minimize_scalar(func)
result
```

```python
# Display per result item

print('x : ' , result.x)
print('y : ' , result.fun)
```

---

Conclusion :

**Function $y = f(x) = x^3 - 3x + 1$ has local/global minimum at point (1, -1).**

```python
# Plot of the function

x = np.linspace(-3, 3, 100)
y = x**3 - 3*x + 1

plt.plot(x, y, label = 'function')
plt.plot(x, [-1]*100, label = 'local/global minimum', linestyle = '--')
plt.legend()
```

### C.2 - Find Local/Global Maximum

Scipy only has function to find minimum points. To find the maximum points, we can multiply the function by `-1`.

```python
# Find the local/global maximum points

## Import library
from scipy.optimize import minimize_scalar

## Define the function
def func(x):
  return -1 * (x**3 - 3*x + 1)

## Display the result
result = minimize_scalar(func)
result
```

As seen above, the `y` value is at `-3.0`. To get the value of `y` where there is a maximum point, multiply this result by `-1`.

```python
# Display the global maximum

print('Global maximum')
print('x : ', result.x)
print('y : ', minimize_scalar(func).fun * (-1))
```

## D. Partial Derivative

Find the partial derivative of $x$ and $y$ from a function $f(x,y) = x^2 + 5xy + 2y^2
$

```python
# Define the symbols
x, y = sy.symbols('x y')

# Define the function
function = x**2 + 5*x*y + 2*y**2
function
```

```python
# Partial derivative with respect of x (First order : fx(x, y))

derivative_x = sy.diff(function, x)
derivative_x
```

```python
# Partial derivative with respect of y (First order : fy(x, y))

derivative_y = sy.diff(function, y)
derivative_y
```

```python
# Second order derivatives - Syntax 1

derivative_x_x = sy.diff(derivative_x, x)   # fxx(x, y)
derivative_x_y = sy.diff(derivative_x, y)   # fxy(x, y)
derivative_y_x = sy.diff(derivative_y, x)   # fyx(x, y)
derivative_y_y = sy.diff(derivative_y, y)   # fyy(x, y)

print('Second order derivative fxx(x, y) : ', derivative_x_x)
print('Second order derivative fxy(x, y) : ', derivative_x_y)
print('Second order derivative fyx(x, y) : ', derivative_y_x)
print('Second order derivative fyy(x, y) : ', derivative_y_y)
```

```python
# Second order derivatives - Syntax 2

print('Second order derivative fxx(x, y) : ', function.diff(x).diff(x))
print('Second order derivative fxy(x, y) : ', function.diff(x).diff(y))
print('Second order derivative fyx(x, y) : ', function.diff(y).diff(x))
print('Second order derivative fyy(x, y) : ', function.diff(y).diff(y))
```

## E. Jacobian Matrix

Example 1 : $f(x, y) = x^2 + 3y$

```python
# Define the equation

matrix_1 = sy.Matrix([x**2 + 3*y ])
matrix_1
```

```python
# Display the Jacobian Matrix

## Defining the mathematical symbols to look for its derivatives
variables = sy.Matrix([x, y])

## Search the Jacobian matrix
matrix_1.jacobian(variables)
```

Example 2 : $f(x, y) = (x^4 + 3xy^2, 5y^2 - 2xy + 1)$

```python
# Define the equation

matrix_2 = sy.Matrix([x**4 + 3*x*y**2, 5*y**2 - 2*x*y + 1])
matrix_2
```

```python
# Display the Jacobian Matrix

## Defining the mathematical symbols to look for its derivatives
variables = sy.Matrix([x, y])

## Search the Jacobian matrix
matrix_2.jacobian(variables)
```

## F. Hessian Matrix

## G. Use Case : Edge Detection

```python
# Import Libraries

import urllib.request
from PIL import Image
```

```python
# Define the image

url_image = 'https://media.istockphoto.com/id/1167975236/photo/the-statue-of-liberty-over-the-scene-of-new-york-cityscape-river-side-which-location-is-lower.webp?b=1&s=170667a&w=0&k=20&c=TeQDUU-lUjpgoJi-r1cvorzsQ8aHPrvsd_Q4pDku4oY='
filename = 'monas.jpg'
```

```python
# Download and load the image

urllib.request.urlretrieve(url_image, filename)
img = plt.imread(filename)
plt.imshow(img)

print('Shape of image : ', img.shape)
```

```python
# Get the red channel

img_red_channel = img[:, :, 0]
plt.imshow(img_red_channel, cmap='gray')

print('Shape of image : ', img_red_channel.shape)
```

```python
# Get the derivative with respect of x and y

img_red_channel_dx, img_red_channel_dy = np.gradient(img_red_channel)

fig, ax = plt.subplots(ncols=3,figsize=(21,5))

ax[0].imshow(img_red_channel, cmap='gray')
ax[0].set_title('Red channel of image')

ax[1].imshow(img_red_channel_dx, cmap='gray')
ax[1].set_title('Derivative of x')

ax[2].imshow(img_red_channel_dy, cmap='gray')
ax[2].set_title('Derivative of y')
```

Gradient can see the texture of an image.

```python
# Show the edge of an image

plt.imshow(np.sqrt(img_red_channel_dx**2 + img_red_channel_dy**2), cmap='gray')
```
