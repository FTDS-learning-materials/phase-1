### D.2 - Feature Creation (Handling High Cardinality)

When working with transactional data, high-cardinality categorical features—such as merchant IDs, user locations, and payment methods—can be challenging to incorporate into machine learning models due to their large number of unique values. Simply encoding them using one-hot encoding or label encoding can lead to sparse data and overfitting. Instead, we can derive meaningful aggregated features that capture user behavior patterns, which can help in predicting churn.

To effectively handle high-cardinality columns, we will:

- Transform raw categorical data into user-level behavioral features

- Extract insights that are interpretable and relevant for churn modeling

- Aggregate high-cardinality values into fewer meaningful metrics

- By creating these engineered features, we can reduce dimensionality, improve model performance, and gain deeper insights into user behavior.

Now, let's walk through each feature and explain how it contributes to churn prediction.

#### `user_id`

> Make total transactions per user. Users with low transaction counts may indicate potential churn.

```python
# Create `user_transaction`

user_transactions = df.groupby('user_id')['transaction_id'].count().reset_index()
user_transactions
```

```python
# Rename the `transaction_id` column to `user_transactions`

user_transactions = user_transactions.rename(columns={'transaction_id': 'user_transactions'})
df = df.merge(user_transactions, on='user_id', how='left') # Merge the dataframes based on user_id, use left to make sure that all data in df exist eventhough there is no transaction
df
```

If a user’s total transactions decrease over time, they might be at risk of churn.

#### `geo_location`

> In geo location we can extract latitude and longitude data. We can calculate the geographical variability of user transactions by finding the standard deviation of its coordinates. This could indicate if users transacting only in one city might churn if a competitor opens there. Users with low geo-variation may be dependent on local merchants.

```python
# Create standard deviation from user coordinates

df[['latitude', 'longitude']] = df['geo_location'].str.split(', ', expand=True).astype(float)
user_dist_var = df.groupby('user_id')[['latitude', 'longitude']].std().reset_index()
user_dist_var.columns = ['user_id', 'latitude_std', 'longitude_std']
user_dist_var
```

```python
# Merge the calculated location variability features back into original dataframe

df = df.merge(user_dist_var, on='user_id', how='left')
df
```

#### `user_agent`

> Make new column called `device_type`. This column will categorize the type of device used in each transaction based on information found in `user_agent` column.

```python
# Create column `device_type

df['device_type'] = np.where(
  df['user_agent'].str.contains('Mac', case=False, na=False), 'Apple',
  np.where(df['user_agent'].str.contains('Android', case=False, na=False), 'Android', 'Other'))

df
```

## E. Task3: Generate Target Variable

Churn is defined as a customer who hasn't made a transaction in the past 90 days. To determine churn status, we wil :

1. **Calculate Transaction Recency**

  Calculate the number of days between each transaction date and the analysis date (`2024-06-01`).

2. **Group by user**
  
  Next, group the transactions by `user_id` to analyze each user individually.

3. **Identify maximum recency**
  
  Within each user group, find the maximum recency. This represents the number of days since the user's most recent transaction relative to the analysis date (`2024-06-01`).

4. **Assign churn label**
   
  If a user's maximum recency is greater than 90 days, they are considered churned (`1`). Otherwise, they are considered as active user (`0`).

```python
# Calculate days since last transaction for each transaction
analysis_date = datetime(2024, 6, 1)
df['recency'] = (analysis_date - df['transaction_date']).dt.days

# Define churn based on maximum days since last transaction for each user
df['churn'] = df.groupby('user_id')['recency'].transform('max')
df['churn'] = (df['churn'] > 90).astype(int)  # 1 if churned, 0 otherwise
df
```

```python
# Print churn statistics
churned_customers_count = df['churn'].sum()
print(f"\nNumber of churned customers: {churned_customers_count}")

unique_churned_customers = df[df['churn'] == 1]['user_id'].nunique()
print(f"Number of unique churned customers: {unique_churned_customers}")
```
